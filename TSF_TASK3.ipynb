```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# **Data Cleaning**
## checking for null values and duplicates

```python
df.isnull().sum()
```


## Checking For duplicate values and dropping those duplicates from the dataset

```python
df[df.duplicated()]
```

```python
len(df[df.duplicated()])
```

```python
df1=df.drop_duplicates()
df1.shape
```

# Dropping Columns that are not useful for the Exploratory Data Analysis

```python
df1=df1.drop(['Country'], axis = 1)
df1=df1.drop(['Postal Code'], axis = 1)
```

# Checking for Correaltion 

```python

sns.heatmap(df1.corr(), cmap = 'viridis', annot = True)
```

```python
df1.head()
```

# Data Visualization

## No of Orders with respect to  Category using countplot

```python
sns.countplot(df1.Category)
```

# Finding total of Measures like profit ,Sales and Quantity

```python
print('Total sales :',round(sum(df1['Sales']), 2))
print('Total Quantity sold :',round(sum(df1['Quantity'])))
print('Total Profit ($) :',round(sum(df1['Profit']), 2))
```

# No of Orders with respect to Segment  using countplot

```python

sns.countplot(df1['Segment'], palette = 'PuBu')
plt.tight_layout()
```

# No of Orders with respect to Region using countplot

```python

sns.countplot(df1['Region'], palette = 'coolwarm')
plt.tight_layout()
```

# No of Orders with respect to Ship Mode using countplot

```python
sns.countplot(df1['Ship Mode'], palette = 'coolwarm')
```

# Orders  with respect to Sub - Categories 

```python
sns.countplot(x=df1['Sub-Category'])
plt.xticks(rotation=60)
```

# Sales With respect to Category

```python
plt.figure(figsize=(10,7))
plt.bar('Category','Sales',data=df1)
plt.title('Category vs Sales')
plt.xlabel('Category')
plt.ylabel('Sales($)')
```

# Profit with respect to Category

```python
plt.figure(figsize=(10,7))
plt.bar('Category','Profit',data=df1)
plt.title('Category vs Profit')
plt.xlabel('Category')
plt.ylabel('Profit')
```

# Average Region wise Sales 

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['Region'])['Sales'].agg('mean').reset_index()
vizrs=sns.barplot(x = 'Region', y='Sales', data=state1,palette='viridis')
vizrs.set_xticklabels(vizrs.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('Region', fontsize=20) 
plt.ylabel('Average Sales', fontsize=20)
```

# Average Region wise Profit

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['Region'])['Profit'].agg('mean').reset_index()
vizr=sns.barplot(x = 'Region', y='Profit', data=state1,palette='viridis')
vizr.set_xticklabels(vizr.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('Region', fontsize=20) 
plt.ylabel('Average Profit', fontsize=20)
```

# **No of Orders with respect to State**

```python
plt.figure(figsize=(10,7))
sns.countplot(x=df1['State'],palette='coolwarm',order=df1['State'].value_counts().index)
plt.xlabel('State', fontsize=20) 
plt.ylabel('Count', fontsize=20)
plt.xticks(rotation=90,fontsize=10)
```

# **Comparing State wise Average Profit**

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['State'])['Profit'].agg('mean').reset_index()
viz1=sns.barplot(x = 'State', y='Profit', data=state1,palette='coolwarm')
viz1.set_xticklabels(viz1.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('State', fontsize=20) 
plt.ylabel('Average Profit', fontsize=20)
```

# **Comparing State wise Average Sales**

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['State'])['Sales'].agg('mean').reset_index()
viz2=sns.barplot(x = 'State', y='Sales', data=state1,palette='viridis')
viz2.set_xticklabels(viz2.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('State', fontsize=20) 
plt.ylabel('Average Sales', fontsize=20)
```

# **Comparing Sub -Category  wise Average Sales**

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['Sub-Category'])['Sales'].agg('mean').reset_index()
viz2=sns.barplot(x = 'Sub-Category', y='Sales', data=state1,palette='viridis')
viz2.set_xticklabels(viz2.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('Sub-Category', fontsize=20) 
plt.ylabel('Average Sales', fontsize=20)
```

# **Comparing Sub -Category  wise Average Profit**

```python
plt.figure(figsize=(10,8))
state1 = df1.groupby(['Sub-Category'])['Profit'].agg('mean').reset_index()
viz2=sns.barplot(x = 'Sub-Category', y='Profit', data=state1,palette='viridis')
viz2.set_xticklabels(viz2.get_xticklabels(),rotation=90)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('Sub-Category', fontsize=20) 
plt.ylabel('Average Profit', fontsize=20)
```

# **Insights and Conclusion After EDA**
## **With respect to Category :** No of Orders for Office Supplies is the most , But the Sales and Profit of Technology category is the highest. Thus attention must be given to office supplies to increase its profits.

## **With respect to Region :** More orders in the west region also average profit is higher in the west region , South region can be improved as orders are less in that region

## **With respect to Ship mode:** Standard class Ship Mode has highest orders, Same Day Ship mode has lowesr number of orders so we need to focus in this mode maybe giving discounts may help for the improvement of this mode.

## **With respect to State:** California State has highest orders but it doesn't have highest profits so we need to make sure that we make  profits from this state. Vermont and Wyoming has high average sales but less orders so it needs to be increased. Vermont even has the highest average profits.Arizona,Texas,Tennessee,,North Carolina,illinois,Florida,Colorado  we need to pay more attention to these states so that no loss occurs in these states

## **With respect to sub-categories**:Binders and papers have highest orders but low profits,Average sales and profit of copiers is the highest. Other sub-categories like enevlopes , art etc need improvements in sales.
